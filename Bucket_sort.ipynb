```python
def bucket_sort(arr):
  if not arr:
    return arr
  n = len(arr)
  buckets = [[ ] for i in range(n)]
  for num in arr:
    index = int(n*num)
    if index == n:
      index = n-1
    buckets[index].append(num)
  for bucket in buckets:
    bucket.sort()
  result = []
  for bucket in buckets:
    result.extend(bucket)
  return result

arr = [0.78,0.17,0.39,0.26,0.72,0.94,0.21,0.12,0.23,0.68]
result = bucket_sort(arr)
print(result)
```
