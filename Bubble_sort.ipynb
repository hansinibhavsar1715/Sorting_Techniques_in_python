```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(n-1-i):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1] = arr[j+1],arr[j]

arr = [64,34,25,12,22,11,90]
print("Before Sorting: ")
print(arr)
bubble_sort(arr)
print("After Sorting: ")
print(arr)
```

```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    swapped = False
    for j in range(n-1-i):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1] = arr[j+1],arr[j]
        swapped = True
    if not swapped:
      break
arr = [64,34,25,12,22,11,90]
print("Before Sorting: ")
print(arr)
bubble_sort(arr)
print("After Sorting: ")
print(arr)
```
