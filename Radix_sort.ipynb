```python
def counting_sort_exp(arr,exp):
  n = len(arr)
  output = [0]*n
  count = [0]*10
  for i in range(n):
    index = arr[i]//exp
    count[index % 10]+=1
  for i in range(1,10):
    count[i]+=count[i-1]
  i = n-1
  while i>= 0:
    index = arr[i]//exp
    output [count[index % 10]-1] = arr[i]
    count[index % 10] -=1
    i -= 1
  for i in range(n):
    arr[i] = output[i]

def radix_sort(arr):
  max_val = max(arr)
  exp = 1
  while max_val // exp>0:
    counting_sort_exp(arr,exp)
    exp *= 10

arr = [170,45,75,90,802,24,2,66]
radix_sort(arr)
print("Sorted: ",arr)
```
